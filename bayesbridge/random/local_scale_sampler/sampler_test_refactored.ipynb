```python
import sys
sys.path.append('.')

import numpy as np
import matplotlib.pyplot as plt

from skewed_local_sampler import skewed_shrinkage_rejection_sampler_sim, sample_c_positive, compute_inflection_point
from helper import plot_hist_against_target, remove_figure_box_edges, plot_hist_against_orig_target


# skewed_shrinkage_rejection_sampler(a: float, c: float, Nsim: int, q: float, k1: int, k2: int)
```

```python
a_c_list = [(0.3,6), (1,1), (10,-10)]
lscale_samples = [
    skewed_shrinkage_rejection_sampler_sim(a = a_c[0], c = a_c[1], nsim = 10 ** 5, q = 0.5, k1 = 16, k2 = 5)
    for a_c in a_c_list
]
```

```python
plt.rcParams['font.size'] = 14
fig, axes = plt.subplots(1, 3, figsize=(15, 4.25))

for i in range(len(a_c_list)):
    plot_hist_against_target(
        axes[i], lscale_samples[i][0], a = a_c_list[i][0], c = a_c_list[i][1]
    )
    remove_figure_box_edges(axes[i])

axes[0].legend(frameon=False)
plt.tight_layout()
plt.show()
```

```python
plt.rcParams['font.size'] = 14
fig, axes = plt.subplots(1, 3, figsize=(15, 4.25))

for i in range(len(a_c_list)):
    plot_hist_against_orig_target(
        axes[i], lscale_samples[i][0], a = a_c_list[i][0], c = a_c_list[i][1]
    )
    remove_figure_box_edges(axes[i])

axes[0].legend(frameon=False)
plt.tight_layout()
plt.show()
```
